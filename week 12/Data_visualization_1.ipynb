```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.arange(0,10)
y = x^2

plt.title("Sample plot")
plt.xlabel("Time")
plt.ylabel("Distance")

plt.plot(x,y, 'r')
plt.plot(x,y, '>')
plt.savefig("TimeDistancePlot.pdf", format="pdf")
```

```python
#Creating a pie chart with matplotlib.pyplot
labels = 'Cow', 'Frog', 'Dog', 'Goat'
sizes = [45, 5, 20, 30]
explode = (0.1,0,0,0)
ax1 = plt.subplots()
plt.pie(sizes,explode=explode,labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```
