```python
def parse_xml(xml_data, df):
    # Initializing soup variable
    soup = BeautifulSoup(xml_data, 'xml')

    # Iterating through exchange tag and extracting elements
    all_items = soup.find_all('exchange')
    #print(all_items)
    items_length = len(all_items)

    rows = []
   
    for index, item in enumerate(all_items):
        dataset = item.find('exchange', {'datasetinternalid'})
       
        short_desc = item.find_all('common:shortDescription')
        epd_amount_array = item.find_all('epd:amount')
        exchange_direction=item.find_all('exchangeDirection')
        if len(exchange_direction) > 0:
            exchange_direction = exchange_direction[0].text
        #functional_unit = item.find_all('common:shortDescription')
        if len(short_desc) > 2:
            functional_unit = short_desc[2].text
            short_description = short_desc[0].text
        elif len(short_desc) == 2:
            functional_unit = short_desc[1].text
            short_description = short_desc[0].text
        elif len(short_desc) == 1:
            functional_unit = short_desc[0].text
            short_description = ""
        else:
            functional_unit = ""
            short_description = ""
        
#         functional_unit = item.find('epd:referenceToUnitGroupDataSet').find('common:shortDescription', {'xml:lang': 'de'})
        
#         exchange_direction=item.find('exchangeDirection')
#         functional_unit=item.find_all('common:shortdescription')
#         if functional_unit.count('common:shortdescription') > 1:
#             functional_unit=functional_unit[1].get_text()
#         else:
#             functional_unit=""
        epd_obj = {}
        for epd_item in epd_amount_array:
            epd_obj[epd_item.get('epd:module')] = epd_item.text 

        # Adding extracted elements to rows in table
        row = {
            'exchange dataSetInternalID': dataset,
            
            'common:shortDescription': short_description,
            'exchange_direction':  exchange_direction,
            'functional_unit':functional_unit
        }
        row.update(epd_obj)

        rows.append(row)
        
    # Iterating through LCIAResult tag and extracting elements
    all_items = soup.find_all('LCIAResult')
    #print(all_items)
    items_length = len(all_items)
    for index, item in enumerate(all_items):
        dataset = ""
        
        short_desc = item.find_all('common:shortDescription')
        epd_amount_array = item.find_all('epd:amount')
        exchange_direction=item.find_all('exchangeDirection')
        if len(exchange_direction) > 0:
            exchange_direction = exchange_direction[0].text
        #functional_unit = item.find_all('common:shortDescription')
        if len(short_desc) > 2:
            functional_unit = short_desc[2].text
            short_description = short_desc[0].text
        elif len(short_desc) == 2:
            functional_unit = short_desc[1].text
            short_description = short_desc[0].text
        elif len(short_desc) == 1:
            functional_unit = short_desc[0].text
            short_description = ""
        else:
            functional_unit = ""
            short_description = ""
        
#         functional_unit = item.find('epd:referenceToUnitGroupDataSet').find('common:shortDescription', {'xml:lang': 'de'})
        
#         exchange_direction=item.find('exchangeDirection')
#         functional_unit=item.find_all('common:shortdescription')
#         if functional_unit.count('common:shortdescription') > 1:
#             functional_unit=functional_unit[1].get_text()
#         else:
#             functional_unit=""
        epd_obj = {}
        for epd_item in epd_amount_array:
            epd_obj[epd_item.get('epd:module')] = epd_item.text 

        # Adding extracted elements to rows in table
        row = {
            'exchange dataSetInternalID': dataset,
            'common:shortDescription': short_description,
            'exchange_direction':  exchange_direction,
            'functional_unit':functional_unit
        }
        row.update(epd_obj)

        rows.append(row)
        print(f'Appending row %s of %s' % (index+1, items_length))

    return pd.concat([df, pd.DataFrame(rows)])

df = pd.DataFrame(columns=['exchange dataSetInternalID','common:shortDescription','exchange_direction','functional_unit', 'A1-A3', 'A4', 'A5', 'B1', 'B2', 'B3', 'B4', 'B5', 'B6',
                               'C1', 'C2', 'C3', 'C4', 'D'])

for root, dirs, files in os.walk("C:/Users/hp/Download/MLEPD/MLEPD"):
    for filename in files:
        if filename.endswith('.xml'):
            with open(os.path.join(root, filename), 'r', encoding='utf-8', errors='ignore') as f:
                file = f.read()
                df = parse_xml(file, df)
df.to_csv('combine4.csv')
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
global_warming_df.isnull().sum()
```

```python
global_warming_df=global_warming_df.dropna(subset=['C2'])
```

```python
global_warming_df.isnull().sum()
```

```python
# Create target object and call it y
y = global_warming_df.A1A3

# Create X
features = ['C2', 'C4', 'D']
X = global_warming_df[features]
```

```python
# Split into validation and training data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
# Specify Model
model = DecisionTreeRegressor(random_state=1)
# Fit Model
model.fit(train_X, train_y)
```

```python
# Make validation predictions and calculate mean absolute error
val_predictions = model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
print("Validation MAE when not specifying max_leaf_nodes: {:,.0f}".format(val_mae))
```

```python
# Using best value for max_leaf_nodes
model = DecisionTreeRegressor(max_leaf_nodes=100, random_state=1)
model.fit(train_X, train_y)
val_predictions = model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
print("Validation MAE for best value of max_leaf_nodes: {:,.0f}".format(val_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
# Split the data into training and validation sets
train_df, val_df = train_test_split(global_warming_df, test_size=0.2, random_state=42)
```

```python
# Define the features and target variable
features = features
target = 'A1A3'

# Create the feature matrix and target array for training and validation
train_X, train_y = train_df[features], train_df[target]
val_X, val_y = val_df[features], val_df[target]

# Instantiate the random forest regressor with desired hyperparameters
rf = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)

# Fit the random forest model on the training data
rf.fit(train_X, train_y)
```

```python
#Predict on the validation set and calculate R-squared score
val_predictions = rf.predict(val_X)
val_r2_score = rf.score(val_X, val_y)
print(f"Validation set R-squared score: {val_r2_score:.4f}")
```

```python
#Predict on the test set and calculate R-squared score
test_predictions = rf.predict(test_X)
test_r2_score = rf.score(test_X, test_y)
print(f"Test set R-squared score: {test_r2_score:.4f}")
```

```python
#Calculate root mean squared error on the validation set
val_rmse = mean_squared_error(val_y, val_predictions, squared=False)
print(f"Validation set Root Mean Squared Error: {val_rmse:.2f}")
```

```python
#Calculate root mean squared error on the test set
test_rmse = mean_squared_error(test_y, test_predictions, squared=False)
print(f"Test set Root Mean Squared Error: {test_rmse:.2f}")
```

```python
#Print the accuracy score of the random forest model on the validation set
accuracy_score = rf.score(val_X, val_y)
print(f"Random Forest Model Accuracy on the validation set: {accuracy_score:.2f}")
```

```python
#Print the accuracy score of the random forest model on the test set
test_accuracy_score = rf.score(test_X, test_y)
print(f"Random Forest Model Accuracy on the test set: {test_accuracy_score:.2f}")
```
